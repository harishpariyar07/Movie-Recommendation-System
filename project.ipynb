```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.head(1)
```

```python
# We only keep the following columns as they are relevant to our usecase
# genres
# id
# keywords
# original_title
# overview
# popularity
# vote_count
# vote_average
# cast
# crew

movies = movies[['genres','id','keywords','original_title','overview', 'popularity', 'vote_count', 'vote_average', 'cast', 'crew']]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
# The duplicate columns are removed and the data frame is updated
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# function that converts the individual objects to list and add up the name property of every list
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
# function that extracts the top 3 casts from every movie
import ast
def convert1(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert1)
```

```python
movies.head(1)
```

```python
# function that extracts only the director from the crew
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
# convert the overview string into a list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
# removing spaces between the words to make comparsion easier
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head(1)
```

```python
# combining overview, genres, keywords,
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
# creating a new data frame with the updated fields
new_df = movies[['id', 'original_title', 'tags', 'popularity', 'vote_count', 'vote_average']]
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# adding a space between elements of tags field
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
# make the string of 'tags' field lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    print(y)
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# similarity of a movie with itself is 1
similarity = cosine_similarity(vectors)
```

```python
# sort the movies in descending order
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['original_title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        #to select a specific row or column from the data set. 
        print(new_df.iloc[i[0]].original_title)
    return 
```

```python
recommend('Avatar')
```
